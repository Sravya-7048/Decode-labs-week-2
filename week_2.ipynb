```python
import pandas as pd

df = pd.read_excel('/content/Dataset for Data Analytics.xlsx')
display(df.head())
```

2. Check Data Quality First

```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

3. Handle Missing Values

```python
df.isnull().sum()
```

```python
df["CouponCode"].fillna("No Coupon", inplace=True)
```

```python
df.fillna(method='ffill', inplace=True)  # forward fill
```

4. Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

5. Fix Data Types

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Quantity"] = df["Quantity"].astype(int)
df["UnitPrice"] = df["UnitPrice"].astype(float)
df["TotalPrice"] = df["TotalPrice"].astype(float)
```

6. Clean Text Columns

```python
text_cols = ["Product", "PaymentMethod", "OrderStatus", "ReferralSource"]

for col in text_cols:
    df[col] = df[col].str.strip()
```

7. Standardize Categories

```python
df["OrderStatus"] = df["OrderStatus"].str.title()
df["PaymentMethod"] = df["PaymentMethod"].str.title()
df["ReferralSource"] = df["ReferralSource"].str.title()
```

8. Validate Calculations

```python
df["CalculatedTotal"] = df["Quantity"] * df["UnitPrice"]

# Check mismatches
df[df["TotalPrice"] != df["CalculatedTotal"]]
```

9. Drop Unnecessary Columns

```python
df.drop(columns=["TrackingNumber"], inplace=True)
```

10. Save Clean Dataset

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)
```
